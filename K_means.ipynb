# K-Means Clustering

K-means is an **unsupervised** algorithm used for **clustering**.

It partitions a dataset into K distinct clusters based on feature similarity. 

### How does K-Means work?

The first step is to decide the *number of clusters* (K) the data is going to be divided into. Then the *cluster centroids* are initialized. This is typically done randomly, but there are smarter initialization techniques (like K-means++). These centroids are the initial "representative points" of each cluster.

Next step is to assign every data point in the dataset to the cluster whose centroid is closest, based on a *distance metric* (e.g., Euclidean distance). Once this is done, a new centroid is computed for each cluster by taking the mean of all data points assigned to that cluster.

The last 2 steps (the datapoints assignment and the new centroid selection) are repeated until the centroids stop changing significantly or the data point assignments remain constant (in other words, the algorithm converges), or a *predefined number of iterations* is reached.

### Pros and cons of the K-Means algorithm

K-Means is simple and very easy to implement. For this reason is one of the most popular clustering techniques. It is efficient for large datasets as well.

On the downside, deceiding the number of clusters behorehand can be challenging. Another thing to keep in mind is that the algorithm is sensitive to the initial placement of centroids.

Lastly, K-Means works very good when clusters have roughly a spherical shape, which may not be the case in all datasets.

Let's play with some artificially generated data:

```python
# Let's import the necessary libraries and generate some synthetic data to start working on

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans


# Generate synthetic dataset with 4 clusters
X, y = make_blobs(n_samples=5000, centers=[[4,4], [-2, -1], [2, -3], [1, 1]], random_state=42, cluster_std=0.9)

# Visualize the data
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title("Generated Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
# Apply KMeans clustering
k = 4  # Number of clusters
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X)
```

```python
# Get cluster assignments and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
# Plot the clusters with their centroids
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis', alpha=0.6)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, marker='X', label='Centroids')
plt.title(f"K-Means Clustering with {k} Clusters")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```
